```python
df= pd.read_csv("/kaggle/input/tmdb-movie-metadata/tmdb_5000_movies.csv")
df.head()
```

```python
df.revenue.describe()
```

```python
df['rev_new']= df['revenue'].apply(lambda x: x/1000000)
df.rev_new.describe()
```

```python
df.head(2)
```

```python
_,mean,std,*_ = df.rev_new.describe()
```

```python
mean,std
```

```python
def get_z_score(value,mean,std):
    return (value-mean)/std
```

```python
df['z_score']= df.rev_new.apply(lambda x: get_z_score(x,mean,std))
df.head(3)
```

```python
df[df.z_score>4.902543]
```
